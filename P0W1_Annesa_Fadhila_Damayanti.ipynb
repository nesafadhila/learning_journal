```python
pokemon = df.copy()
```

```python
pokemon.rename(columns= {'Type 1' : 'Type_1', 'Type 2' : 'Type_2' , 'Sp. Atk' : 'Sp_Atk', 'Sp. Def' : 'Sp_Def'}, inplace=True)
```

```python
pokemon.head(2)
```

```python
pokemon.drop(['Generation', 'Legendary'], inplace=True, axis=1)
```

```python
pokemon.head(2)
```

```python
pokemon.shape
```

```python
pokemon['Difference'] = np.abs(pokemon['Attack'] - pokemon['Defense'])
```

```python
pokemon.sample(3)
```

```python
pokemon.query('Type_2 == "Water" & HP > 40 & Speed > 20')
```

```python
pokemon.query('Type_1 == "Ice" & Attack < 50 & Speed > 30')
```

```python
pokemon.query('Attack > 100 & Defense < 100 & Speed > 20 & Difference < 35')
```

```python
pokemon.query('Type_2 == "Flying" & HP > 50 & Defense < 100')
```

```python
pokemon.query('Type_2 == "Rock" & HP < 70 & Attack > 50 & Speed > 20')
```

```python
pokemon.groupby('Name')['HP'].sum()
```

```python
pokemon.groupby('Name')['Type_1','Type_2'].min()
```

```python
pokemon.groupby('Name') ['Attack','Defense','Difference'].max()
```

```python
pokemon[(pokemon['Type_1'] == 'Rock') &
        (pokemon['Attack'] > 40) &
        (pokemon['Defense'] < 80) &
        (pokemon['Difference'] > 10)
].groupby(['Name']).mean()
```

```python
pokemon[(pokemon['Type_2'] == 'Flying') &
        (pokemon['Attack'] > 50)
].groupby(['Name']).mean()
```

```python
%matplotlib inline
```

```python
pokemon[pokemon["Type_1"] == 'Dark'].groupby(['Attack','Defense']).sum().plot()
```

```python
pokemon[pokemon['Type_2']== 'Ice'].head(5).plot(kind="bar")
```

```python
pokemon[pokemon['Speed'] > 30].groupby(['Name','Type_1','Type_2']).head(10).mean().plot(kind='pie')
```
