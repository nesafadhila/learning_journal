```python
grey = 0.299*Brown + 0.587*Green + 0.114*Yellow
grey.shape
```

```python
fig,ax = plt.subplots(ncols=2,figsize=(15,10))
ax[0].imshow(grey)
ax[0].set_title('grayscale tanpa cmap')
ax[1].imshow(grey, cmap='gray')
ax[1].set_title('grayscale dengan cmap')
```

```python
grey = 0.299*Brown + 0.587*Green + 0.114*Yellow
dx, dy = np.gradient(grey)
fig,ax = plt.subplots(ncols=3,figsize=(20,10))
ax[0].imshow(grey, cmap= 'gray')
ax[0].set_title('grayscale')
ax[1].imshow(dx, cmap='gray')
ax[1].set_title(r'$\partial_x img$')
ax[2].imshow(dy, cmap= 'gray')
ax[2].set_title(r'$\partial_y img$')
```

```python
mag = (dx*dx+dy*dy)**(1/2)
mag
```

```python
mag1 = (mag > 30)
```

```python
mag2 = (mag > 100)
```

```python
mag3 = (mag > 150)
```

```python
fig, ax = plt.subplots(ncols=5,figsize=(25,15))
ax[0].imshow(grey, cmap='gray')
ax[0].set_title('Greyscale')
ax[1].imshow(mag)
ax[1].set_title('Magnitue')
ax[2].imshow(mag1)
ax[2].set_title('Threshold 30')
ax[3].imshow(mag2)
ax[3].set_title('Threshold 100')
ax[4].imshow(mag3)
ax[4].set_title('Threshold 150')
```

threshold 30 : masih terlihat garis-garis pada gambar, terlihat dengan jelas
threshold 100 : ada beberapa titik yang memperlihatkan garis gambar tertapi buram
threshold 150 : garis pada gambar terlihat buram dan hanya pada di ujung sudut kanan saja
pada gambar-gambar yang ditampilkan terlihat, semakin kecil nilai threshold maka akan semakin terlihat jelas garis gambar. sedangkan semakin besar nilai threshold maka akan semakin tidak terlihat garis gambar.

```python
random = np.random.rand(780, 438)
random
```

```python
X = mag1 @ random
X
```

```python
fig,ax = plt.subplots(ncols=2,figsize=(15,10))
ax[0].imshow(X)
ax[0].set_title('tanpa cmap')
ax[1].imshow(X, cmap = 'gray')
ax[1].set_title('dengan cmap')
```

random matriks yang dikalikan dengan magnitude edge terbaik menghasilkan gambar garis lurus secara horizontal. dengan warna lebih banyak warna gelap dibandingkan dengan warna cerah.
